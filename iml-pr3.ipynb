```python
# Makvana ranju
# Enroll: 216270307105
# practical:3 Write a NumPy program to implement following operation
#• to split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order
#• to stack arrays horizontally (column wise)

import numpy as np # linear algebra

#create the original array
original_array = np.arange(14)

#split the array into three subarrays
subarrays = np.split(original_array,[2,6])

#print the subarraysfor subarray in subarrays:
print(subarrays)
```

```python
#to stack arrays horizontally (column wise)

import numpy as np # linear algebra
# Create two arrays
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

# Stack the arrays horizontally
stacked_array = np.hstack((array1, array2))

# Print the stacked array
print(stacked_array)
```
